```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import sys
import statistics
import time
from tqdm import tqdm
sys.path.append(os.path.dirname(os.getcwd()))

from problems.problem_3.p3_a import number_of_large_inversions_3a
from problems.problem_3.p3_b import number_of_large_inversions_3b
```

```python
def create_text_inputs(size):
    # Write text input file    
    with open('input/test_p3_student_n' + str(size) + '.txt', 'w') as f:
        f.write(str(size))
        f.write('\n')
        f.write(" ".join(str(element) for element in range(1,size+1)))
```

```python
## Get run times 
a_times = []
b_times = []

# List sizes
sizes = [int(i) for i in np.linspace(0, 10, 10)]


# Get run times
for size in tqdm(sizes):
    
    a_times_single_size = []
    b_times_single_size = []
    
    for _ in range(100):
        create_text_inputs(size)

        # time a
        start_time = time.time()
        number_of_large_inversions_3a('input/test_p3_student_n' + str(size) + '.txt', random.randint(0,size)) # run function
        a_times_single_size.append(time.time() - start_time)
    

            
        # time b
        start_time = time.time()
        number_of_large_inversions_3b('input/test_p3_student_n' + str(size) + '.txt', random.randint(0,size)) # run function
        b_times_single_size.append(time.time() - start_time)
        
    
    a_times.append(statistics.median(a_times_single_size))
    b_times.append(statistics.median(b_times_single_size))
```

```python
# Plotting
plt.plot(sizes, a_times, marker='o', label='$O(n^2)$', color='blue')
plt.plot(sizes, b_times, marker='o', label='$O(n \log n)$', color='red')
plt.xlabel('List Size')
plt.ylabel('Time (seconds)')
plt.title('Problem 3: $O(n^2)$ vs $O(n \log n)$')
plt.legend()
plt.savefig(os.path.dirname(os.getcwd()) + '/latex/figures/problem-3a.png')
```
